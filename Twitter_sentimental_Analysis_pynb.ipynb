```
import numpy as np
import pandas as pd
```

```
tweets = [
          "Wow, what a great day today!! #sunshine",
    "I feel sad about the things going on around us. #covid19",
    "I'm really excited to learn Python with @JovianML #zerotopandas",
    "This is a really nice song. #linkinpark",
    "The python programming language is useful for data science",
    "Why do bad things happen to me?",
    "Apple announces the release of the new iPhone 12. Fans are excited.",
    "Spent my day with family!! #happy",
    "Check out my blog post on common string operations in Python. #zerotopandas",
    "Freecodecamp has great coding tutorials. #skillup"
]
```

For Analysis we need to find Question which is to be answered.

So i am going to ask quesion myself and perform analysis on these few lines of tweets.

**Q1.How many tweets does the dataset Contain?**

```
num_of_tweets = len(tweets)
num_of_tweets
```

I have created a list of kind of words: Happy_words and sad_words. I will use these to check if a tweet is happy or sad.

```
Happy_words = ['great','happy','excited','wonderful','amazing','nice','good','best']
```

```
sad_words =['sad','bad','tragic','unhappy','worst']
```

## Performing Analysis
At First I am going to find Happy tweets in my tweets and count it and then try to find sad tweets.

**Q.Determine the number of tweets in the dataset that can be classified as happy.**

```
number_of_happy_tweets = 0
```

```
for tweet in tweets:
    for happy_word in Happy_words:
        if happy_word in tweet:
            number_of_happy_tweets = number_of_happy_tweets + 1
            print(number_of_happy_tweets)
            print(tweet)
```

```
print("Number of Happy Tweets are:",number_of_happy_tweets)
```

**Q..Determine the number of tweets in the dataset that can be classified as sad.**

```
number_of_sad_tweets = 0
for tweet in tweets:
    for sad_word in sad_words:
        if sad_word in tweet:
            number_of_sad_tweets+=1
            print(number_of_sad_tweets,tweet)
```

```
print("Number of Sad Tweets are:",number_of_sad_tweets)
```

**Q: What fraction of the total number of tweets are happy?**

```
Happy_fraction = number_of_happy_tweets/len(tweets)
```

```
print("The Fraction of Happy tweets is: {}".format(Happy_fraction))
```

**Q: What fraction of the total number of tweets are sad?**

```
sad_fraction = number_of_sad_tweets/len(tweets)
```

```
print("The Fraction of sad tweets are: {}".format(sad_fraction))
```

**Q: Calculate the sentiment score, which is defined as the difference between the fraction of happy tweets and the fraction of sad tweets**.

```
sentiment_score = Happy_fraction - sad_fraction
```

```
print("The sentiment score for the given tweets is", sentiment_score)
```
